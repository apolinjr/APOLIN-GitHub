# First- and Second-order All-Pass Filters

### First-order all-pass (IIR) system example:
- Example 5.11

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot  as plt

r = .9
theta = 0
# theta = np.pi

p = r*np.exp(1j*theta)
z = 1/np.conj(p)

P = np.array([p])
Z = np.array([z])
k = -np.conj(p)

b, a = signal.zpk2tf(Z,P,k)
# Defining the system manually:
# b = np.array([-np.conj(p), 1])
# a = np.array([1, -p])

w, H = signal.freqz(b,a,worN=512,whole=True)
# angleH = 180*np.unwrap(np.angle(H))/np.pi
# angleH = 180*np.angle(H)/np.pi
# angleH = np.unwrap(np.angle(H))
angleH = np.angle(H)
w, grdH = signal.group_delay((b,a),whole=True)

fig1, axs = plt.subplots(3,1, sharex= True,figsize=(4,6),constrained_layout=True)
fig1.suptitle('First-order All-pass IIR System')
axs[0].plot(w,20*np.log10(abs(H)),'k')
axs[0].set_ylabel('Gain (dB)')
axs[0].set_ylim([-1, 1])
axs[0].grid(True)

axs[1].plot(w,angleH,'k')
axs[1].set_ylim([-4,4])
# axs[1].set_yticks([-4,-2,0,2,4])
axs[1].grid()
axs[1].set_ylabel('Phase (degrees)')

# axs[2].plot(w/np.pi,grdH)
axs[2].plot(w,grdH,'k')
axs[2].set_ylabel('Grd (samples)')
axs[2].set_xlabel('Radian frequency '+ r'$(\omega)$')
axs[2].grid()
axs[2].set_ylim([0,20])
axs[2].set_xlim([0, 2*np.pi])
axs[2].set_xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[2].set_xticklabels([0, r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']);
```

### Second-order All-pass IIR System
- Example 5.11 (continued)

```python
r = .9
theta = np.pi/4

p1 = r*np.exp(1j*theta)
p2 = np.conj(p1)

z1 = 1/np.conj(p1)
z2 = np.conj(z1)

Z = np.array([z1,z2])
P = np.array([p1,p2])

k = p1*p2

b, a = signal.zpk2tf(Z,P,k)

w, H = signal.freqz(b,a,whole=True)
# angleH = 180*np.unwrap(np.angle(H))/np.pi
# angleH = 180*np.angle(H)/np.pi
# angleH = np.unwrap(np.angle(H))
angleH = np.angle(H)
w, grdH = signal.group_delay((b,a),whole=True)

fig2, axs = plt.subplots(3,1, sharex= True,figsize=(4,6),constrained_layout=True)
fig2.suptitle('Second-order All-pass IIR System')
axs[0].plot(w,20*np.log10(abs(H)),'k')
axs[0].set_ylabel('Gain (dB)')
axs[0].set_ylim([-1, 1])
axs[0].grid(True)

axs[1].plot(w,angleH,'k')
axs[1].set_ylim([-4,4])
# axs[1].set_yticks([-4,-2,0,2,4])
axs[1].grid()
axs[1].set_ylabel('Phase (degrees)')

# axs[2].plot(w/np.pi,grdH)
axs[2].plot(w,grdH,'k')
axs[2].set_ylabel('Grd (samples)')
axs[2].set_xlabel('Radian frequency '+ r'$(\omega)$')
axs[2].grid()
axs[2].set_ylim([0,20])
axs[2].set_xlim([0, 2*np.pi])
axs[2].set_xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[2].set_xticklabels([0, r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']);
```
